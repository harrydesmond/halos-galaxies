```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.load("../../Data/halos_list.npy")
```

```python
data
```

```python
mass = np.log10(data['mvir'])
```

```python
def func(x):
    return np.where(mass < x)[0].size/mass.size
```

```python
%timeit func(9.8)
```

```python
x = np.linspace(np.min(mass), 12, 100)
y = [func(i) for i in x]
```

```python
plt.figure(dpi=120)
plt.plot(x, y)
plt.axvline(9.8, color='red', alpha=0.5, label=r"$\log M/M_{\odot}=9.8$")
plt.xlabel(r"$\log M_c/M_{\odot}$")
plt.ylabel(r"$N(\log M < \log M_c)/N_{\mathrm{tot}}$")
plt.grid()
plt.legend()
plt.tight_layout()
plt.savefig("../../Plots/9_Halos.pdf")
plt.show()
```
