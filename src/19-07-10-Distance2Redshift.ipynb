# Distance2Redshift convertor

```python
import numpy as np
import cosmolopy.distance as cd
import cosmolopy.constants as cc
import scipy.constants as consts
import matplotlib.pyplot as plt

from astropy.io import fits
```

```python
# Set cosmology parameters, consistent with previous work (thesis)
cosmology = {u'h': 0.688, u'omega_lambda_0': 0.705,
             u'omega_k_0': 0.0, u'omega_M_0': 0.295}
cosmo = cd.set_omega_k_0(cosmology)
```

```python
def cosmolopy_redshift(dist):
    # This estimate only works for nearby objects
    guess = cosmology['h']*100*dist/(consts.c/1e3)
    return cd.redshift_d_light(dl=dist, z_guess=guess, **cosmo)
```

```python
# Load the data
survey = astropy.io.fits.open('../BAM/a100springfull.fits')
end = 10359

ra = np.deg2rad(survey[1].data['RAdeg_HI'][:end])
dec = np.deg2rad(survey[1].data['DECdeg_HI'][:end])
dist = survey[1].data['Dist'][:end]
```

```python
# Do the actual work
redshifts = list()
for dl in dist:
    redshifts.append(cosmolopy_redshift(dl))
redshifts = np.array(redshifts)
```

### A histogram for a quick check

```python
plt.figure(dpi=120)
plt.hist(redshifts, bins='auto')
plt.xlabel('Redshift hist for the matched catalog')
plt.savefig('../Plots/redshift_dist.pdf')
plt.show()
```

### Save the results in a numpy array

```python
N = ra.size
matched_catalog = np.zeros(N, dtype={'names':('ra', 'dec', 'redshift'),
                      'formats':('f8', 'f8', 'f8')})
matched_catalog['ra'] = ra
matched_catalog['dec'] = dec
matched_catalog['redshift'] = np.ravel(redshifts)

np.save('../Data/matched_cat.npy', matched_catalog)
```
