```python
import numpy as np
import matplotlib.pyplot as plt
import healpy as hp

from astropy.io import fits
h = 1
```

```python
survey = fits.open("../Data/kcorrect.nearest.petro.z0.10.fits")[1].data
```

```python
mass = survey['MASS']/h**2
Z = survey['Z']
RA = survey['RA']
DEC = survey['DEC']
MAG = survey['ABSMAG'][:, 2]-5*np.log10(h)
```

```python
IDS = np.where(MAG < -19)
mass = mass[IDS]
Z = Z[IDS]
RA = RA[IDS]
DEC = DEC[IDS]
MAG = MAG[IDS]

IDS = np.where(np.logical_and(Z>0.005, Z<0.064))
mass = mass[IDS]
Z = Z[IDS]
RA = RA[IDS]
DEC = DEC[IDS]
MAG = MAG[IDS]

IDS = np.where(np.logical_and(RA>100, RA<300))
mass = mass[IDS]
Z = Z[IDS]
RA = RA[IDS]
DEC = DEC[IDS]
MAG = MAG[IDS]

IDS = np.where(mass>0)
mass = np.log10(mass[IDS])
Z = Z[IDS]
RA = RA[IDS]
DEC = DEC[IDS]
MAG = MAG[IDS]

IDS = np.where(mass>9.8)
mass = mass[IDS]
Z = Z[IDS]
RA = RA[IDS]
DEC = DEC[IDS]
MAG = MAG[IDS]
```
