# Get the catalogue and make some plots

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
catalogue = np.load("../BAM/Catalog_Baryon/Catalog_SMF_alpha0.npy")
```

```python
mh = catalogue[:, 0]
ms = catalogue[:, 2]

ratio = np.power(10, ms)/mh
```

Plotting the $M-M_h$ relation for SMF to test it. I assume the plot below should be averaged in log-bins.

```python
plt.figure(dpi=120)
plt.loglog(mh[:1000], ratio[:1000])
plt.show()
```
